```python
# Import required libraries
import os
import pandas as pd
from pathlib import Path
import hvplot.pandas
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# Import required Data,including Increase and Decrease
DTE ='min'
type = 'etfs'
date = '10-11-2022'
increase = pd.read_csv(os.path.join(Path.cwd(),f'../Data/Increase/{type}-increase-change-in-open-interest-{date}.csv'))
decrease = pd.read_csv(os.path.join(Path.cwd(),f'../Data/Decrease/{type}-decrease-change-in-open-interest-{date}.csv'))

#show dataframe

display(increase.head())
display(decrease.head())
```

```python
# check Missing Data

display(increase.isnull().sum())
display(increase.isnull().sum())
```

```python
# Handle Missing Data

increase = increase.dropna()
decrease = decrease.dropna()
```

```python
#Select the data only equal to minimun DTE (closest to expired).

if DTE == 'min':
    increase_min_DTE = increase[increase['DTE'] == increase['DTE'].min()]
    decrease_min_DTE = decrease[decrease['DTE'] == decrease['DTE'].min()]
else:
    increase_min_DTE = increase[increase['DTE'] <= DTE]
    decrease_min_DTE = decrease[decrease['DTE'] <= DTE]


# show dataframe
display(increase_min_DTE)
display(decrease_min_DTE)
```

```python
#Merge data of increase and decrease dataframe

combine_df = increase_min_DTE.merge(decrease_min_DTE, how= 'outer')

```

```python
#Transfer column 'OI Chg'datatype from str to float.
combine_df['OI Chg'] = combine_df['OI Chg'].str.replace(',', '').str.replace('unch', '0').astype(float)
combine_df.dtypes
```

```python
#Transfer column 'IV'datatype from str to float.
combine_df['IV'] = combine_df['IV'].str.replace('%', '').astype(float)/100
combine_df.dtypes
```

```python
# Sort by this orders Symbol,Type,Strike,Open Int,Volume,OI Chg,IV

combine_sort_df = combine_df.sort_values(['Symbol','Type','Strike','Open Int','Volume','OI Chg','IV'])
```

```python
# Create the function for clean data acquire.
def get_data(date,types,DTE):

        increase = pd.read_csv(os.path.join(Path.cwd(),f'../Data/Increase/{types}-increase-change-in-open-interest-{date}.csv'))
        decrease = pd.read_csv(os.path.join(Path.cwd(),f'../Data/Decrease/{types}-decrease-change-in-open-interest-{date}.csv'))
        
        #Combine DataFram
        combine_df = increase.merge(decrease, how= 'outer')
        
        #Select the DTE
        if DTE == 'min':
            combine_min_DTE = combine_df[combine_df['DTE'] == increase['DTE'].min()]
        else:
            combine_min_DTE = combine_df[combine_df['DTE'] <= DTE]
            
        combine_min_DTE = combine_min_DTE.dropna()
        
        
        #Transfer column 'OI Chg'datatype from str to float.
        combine_min_DTE['OI Chg'] = combine_min_DTE['OI Chg'].str.replace(',', '').str.replace('unch', '0').astype(float)
        
        #Transfer column 'IV'datatype from str to float.
        combine_min_DTE['IV'] = combine_min_DTE['IV'].str.replace('%', '').str.replace(',', '').astype(float)/100
        
        #Sort data
        combine_sort_df = combine_min_DTE.sort_values(['Symbol','Type','Strike','Open Int','Volume','OI Chg','IV'])
        
        return combine_sort_df
        
```
