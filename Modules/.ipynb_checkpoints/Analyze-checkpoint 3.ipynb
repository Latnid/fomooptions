```python
from CleanData import get_data
import pandas as pd 
import hvplot.pandas
from pathlib import Path
```

```python
#Import Data
option_change = get_data('01-11-2023','stocks','min')

#show data
option_change.head()
```

```python
#Create DataFrame for all the required columns
option_change_required = option_change[['Symbol','Type','Strike','DTE','Open Int','OI Chg','Volume','Price','IV']]

#show dataframe
option_change_required.head()
```

```python
#Sorting step one
#Groupby Symbol, then caluate the total Open Int, and make it as a new column
option_change_required['Total Open Int'] = option_change_required\
.groupby('Symbol')['Open Int'].transform('sum')

#show DataFrame
option_change_required.head()
```

```python
#Sorting step one
#Acquire the list of the top 20 Symbols

top_20 = option_change_required.groupby('Symbol').agg({'Open Int':'sum'})\
.sort_values('Open Int',ascending=False).iloc[:20]

#show dataframe
top_20
```

```python
#Sorting step three
#Use groupby and transform to add a column called Total Open Int base on option_change_required
option_change_required['Total Open Int'] = option_change_required\
.groupby('Symbol')['Open Int'].transform('sum')

#Use .isin to filter the rows from top_20.index

option_change_required_top_20 = option_change_required[option_change_required\
.Symbol.isin(top_20.index)]\
.sort_values(by = ['Total Open Int','Symbol','Type','Strike','Open Int',], ascending=False)\
.set_index(['Symbol','Type','Strike'])

#Drop Total Open Int 
#option_change_required_top_20.drop(columns='Total Open Int',inplace=True)

#save it to csv file
option_change_required_top_20.to_csv(Path('../Output/3.csv'))

#show dataframe
option_change_required_top_20.head(10)
```

```python
#Top 20 symbols ranked by total open interest，
#Call / Put Open Interest comparation
option_change_required_top_20.hvplot.bar(
    height = 300,
    width = 900,
    x='Symbol',
    xlabel = 'Symbol by Type',
    y='Open Int',
    ylabel = 'Open Interest',
    by='Type',
    hover_cols = ['Strike','DTE'],
    rot=90,
    yformatter = '%0f',
    title = 'Total Open Interest(Call+Put) Top 20'
)
```

```python
option_change_required_top_20.hvplot.bar(
    y='Total Open Int',
    stacked=False,
    height=300,
    width=900, 
    yformatter='%0f',
    rot=45,
    hover_cols = ['Strike','DTE'],
    xlabel='Tickers by Call and Put',
    ylabel='Open Interests',
    title = 'Tickers Call / Put Open Interests comparation'
)
```

```python
option_change_required_top_20.hvplot.bar(
    y='Volume',
    by='Type',
    stacked=False,
    height=300,
    width=900, 
    yformatter='%0f',
    rot=90,
    hover_cols = ['Strike','DTE'],
    xlabel='Tickers by Call and Put',
    ylabel = 'Volume',
    title = 'Tickers Call / Put Volume'
)
```

```python
#Top 20 Open Interest change
option_change_required_top_20.hvplot.bar(
    y='OI Chg',
    by='Type',
    stacked=False,
    height=300,
    width=900, 
    yformatter='%0f',
    rot=90,
    hover_cols = ['Strike','DTE'],
    xlabel='Tickers by Call and Put',
    ylabel = 'Open Interest Change',
    title = 'Tickers Call / Put Open Interests Change comparation'
)
```
